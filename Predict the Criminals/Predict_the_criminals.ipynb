```python
for i in X.columns:
    y_data = []
    for uniq in X[i].unique():
        y_data.append((X[i] == uniq).sum())
    x_data=list(X[i].unique())
    x_axis = i+' distinct values'
    y_axis = i+' count'
    bar_df = pd.DataFrame({x_axis:list(x_data), y_axis:list(y_data)})
    bar_df.plot(x_axis, y_axis, kind='bar', color='r')
```
