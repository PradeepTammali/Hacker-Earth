```python
import numpy as np  
import pandas as pd  
import matplotlib.pyplot as plt  
%matplotlib inline
```

```python
import os  
```

```python
path = os.getcwd() + '/criminal_train.csv'
```

```python
data = pd.read_csv(path)
```

```python
data.head()
```

```python
criminal = data[data['Criminal'].isin([1])]
non_criminal = data[data['Criminal'].isin([0])]
```

```python
fig, ax = plt.subplots(figsize=(12,8))  
ax.scatter(criminal['VESTR'],criminal['ANALWT_C'], s=50, c='b', marker='o', label='criminal')  
ax.scatter(non_criminal['VESTR'],non_criminal['ANALWT_C'], s=50, c='r', marker='x', label='non criminal')  
ax.legend()  
ax.set_xlabel('criminal')  
ax.set_ylabel('non_criminal') 
```

```python
data_corr = data.corr()
mask = np.zeros_like(data_corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
```

```python
import seaborn as sns
sns.set(style="white")
```

```python
fig, ax = plt.subplots(figsize=(20,15))
cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(data_corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
